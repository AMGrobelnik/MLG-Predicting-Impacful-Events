```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import umap
```

```python
df1 = pd.read_pickle(f"../../data/text/embedded_umap_dim100/events-00001.pkl")
df2 = pd.read_pickle(f"../../data/text/embedded_umap_dim100/events-00002.pkl")
df3 = pd.read_pickle(f"../../data/text/embedded_umap_dim100/events-00003.pkl")
df4 = pd.read_pickle(f"../../data/text/embedded_umap_dim100/events-00004.pkl")
df5 = pd.read_pickle(f"../../data/text/embedded_umap_dim100/events-00005.pkl")
```

```python
pca = PCA(n_components=2)
titles_pca_df1 = pca.fit_transform(df1['summary'].tolist())
titles_pca_df2 = pca.transform(df2['summary'].tolist())
titles_pca_df3 = pca.transform(df3['summary'].tolist())
titles_pca_df4 = pca.transform(df4['summary'].tolist())
titles_pca_df5 = pca.transform(df5['summary'].tolist())

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(titles_pca_df1[:, 0], titles_pca_df1[:, 1], color='yellow', alpha=0.5, label='df1 Titles')
plt.scatter(titles_pca_df2[:, 0], titles_pca_df2[:, 1], color='blue', alpha=0.5, label='df2 Titles')
plt.scatter(titles_pca_df3[:, 0], titles_pca_df3[:, 1], color='red', alpha=0.5, label='df3 Titles')
plt.scatter(titles_pca_df4[:, 0], titles_pca_df4[:, 1], color='green', alpha=0.5, label='df4 Titles')
plt.scatter(titles_pca_df5[:, 0], titles_pca_df5[:, 1], color='pink', alpha=0.5, label='df5 Titles')
plt.title('PCA of Titles from df1 and df2')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.show()
```

```python
umap_embedder = umap.UMAP(n_components=2, random_state=42, verbose=True)
titles_umap_df1 = umap_embedder.fit_transform(df1['summary'].tolist())
titles_umap_df2 = umap_embedder.transform(df2['summary'].tolist())
titles_umap_df3 = umap_embedder.transform(df3['summary'].tolist())
titles_umap_df4 = umap_embedder.transform(df4['summary'].tolist())
titles_umap_df5 = umap_embedder.transform(df5['summary'].tolist())

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(titles_umap_df1[:, 0], titles_umap_df1[:, 1], color='yellow', alpha=0.5, label='df1 Titles')
plt.scatter(titles_umap_df2[:, 0], titles_umap_df2[:, 1], color='blue', alpha=0.5, label='df2 Titles')
plt.scatter(titles_umap_df3[:, 0], titles_umap_df3[:, 1], color='red', alpha=0.5, label='df3 Titles')
plt.scatter(titles_umap_df4[:, 0], titles_umap_df4[:, 1], color='green', alpha=0.5, label='df4 Titles')
plt.scatter(titles_umap_df5[:, 0], titles_umap_df5[:, 1], color='pink', alpha=0.5, label='df5 Titles')
plt.title('PCA of Titles from df1 and df2')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.show()
```
